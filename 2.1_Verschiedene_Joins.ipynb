# Fazit

Bei den verschiedenen Joins hat sich lediglich der Outer-Join als sinnvoll erwiesen. Die anderen Joins haben keine erfolgreiche Ergebnisse aufgezeigt, da sich durch die Joins zum einen die Datengrundlage stark reduziert hat und zum anderen sehr viele Missing Values vorhanden sind. Würden diese gejoint Daten noch bereinigt werden, hätten wir keine vernünftige Datenbasis mehr. Somit haben wir uns dazu entschieden, mit dem Datensatz des Outer-Joins weiterzuarbeiten.
