##Runtime

```python
from typing_extensions import runtime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Annahme: Die Variable "runtime" ist in einem DataFrame namens "df" enthalten
# runtime = df['runtime']
runtime = df['runtime']
q25, q75 = np.percentile(runtime, [25, 75])

bin_width = 2 * (q75 - q25) * len(runtime) ** (-1/3)

bins = round((runtime.max() - runtime.min()) / bin_width)

print("Freedman–Diaconis number of bins:", bins)

plt.hist(runtime, bins=bins)
plt.xlabel('Runtime')
plt.ylabel('Frequency')
plt.title('Histogram of Runtime')
plt.show()
```

Die meisten Filme haben eine Laufzeit von 90 Minuten. Es gibt aber auch viele kürzere Filme <=50 Minuten in unserem Datensatz. Dabei haben wir weniger längere Filme >=120 Minuten.

#n-Grams

Funktion für die am meisten benutzten n-Grams

```python
from sklearn.feature_extraction.text import CountVectorizer

def get_top_n_words(corpus, n=None, ngram_range=(1,1)):
  """Retrieves the most common ngrams from the given text corpus."""
  # initialize an instance of CountVectorizer - a sklearn transformer for counting n-grams
  vec = CountVectorizer(ngram_range=ngram_range).fit(corpus)

  # retrieve the bag of words of the corpus by applying the count vectorizer
  bag_of_words = vec.transform(corpus)

  # CountVectorizer returns a threedimensional array where each column corresponds to a n-gram,
  # each row to an text document in the corpus
  # and each element to how many times that n-gram appeared in that document
  sum_words = bag_of_words.sum(axis=0) # sum across rows -> count how many n-grams in all documents

  # go through all n-grams and create a tuple containing the n-gram and its frequency
  words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]

  # sort the n-grams by frequency ascending
  words_freq = sorted(words_freq, key = lambda x: x[1], reverse=True)

  return words_freq[:n] # take n first elements
```
