Wir haben durch die Bereinigung des Outer-Join Datensatzes festgestellt, dass sich die für unsere Modelle entscheidenen Variablen in dem Datensatz "credits" und "titels" befinden. Dies ermöglicht uns den join der beiden Tabellen über die "id", was zu zuverlässigeren Ergebnissen führt als der join über die anderen Variablen. Somit werden wir hier nochmal die beiden Tabellen über die "id" joinen und die Daten bereinigen.




Unsere bereinigten Daten einlesen:

```python
import pandas as pd
```

```python
credits = pd.read_csv('/content/sample_data/credits_join.csv')
```

```python
titles = pd.read_csv('/content/sample_data/titles_join.csv')
```

##Join 1: Daten "credits" und "titles" joinen

Zu aller erst schauen wir uns die einzelnen beiden Datensätze an

```python
credits
```

```python
titles
```

Joinen der beiden Datensätze

```python
df= pd.merge(credits, titles, on= ['id'], how="outer") #Join1 von credits daten und titles Daten (Beide bereinigt)
df
```

# Bereinigung

Als erstes werden wir die Spalte "id" löschen, da diese für unsere weiteren Schritte nicht mehr von Bedeutung ist:

```python
df= df.drop(['id'], axis=1)
df
```

```python
#prüfen nach fehlenden Werten:

df.isna().sum()
```

```python
#Zunächst löschen wir die fehlenden Werte in der title Spalte und der imdb und tmdb score Spalten, da diese für unsere Modell am wichitgsten sind

df = df.dropna(subset = ['title', 'imdb_score', 'tmdb_score'], axis=0)
df
```

```python
#prüfen nach fehlenden Werten:

df.isna().sum()
```

Wir löschen nun auch die anderen fehlenden Werte, da unser Datensatz immer noch groß genug ist, wenn wir diese herausnehmen:

```python
df=df.dropna()
df
```

```python
#Kontrolle, on noch fehlende Werte vorhanden:

df.isna().sum()
```

Da die Schauspieler der Filme angegeben sind, sind die Filme mehrmals enthalten. Wir wollen nun überprüfen, ob durch das löschen der fehlende Werte wir noch eine ausreichende Datenbasis für unsere Modelle haben. Wir überprüfen also, wie viele Filme im Datensatz enthalten sind:

```python
uniques = pd.unique(df[['title']].values.ravel())
len(uniques)
```

Wir sehen damit, dass wir trotzt löschen der Missing values eine ausreichend große Datenbasis haben. Nun müssen wir noch überprüfen, ob noch Duplikate im Datensatz vorliegen:

```python
df.duplicated().sum()
```

```python
#Da nur 3 duplizierte Zeilen im Datensatz vorhanden sind, werden wir diese entfernen:

df = df.drop_duplicates()
df
```

```python
#Nun haben wir unsere gejointe Datei bereinigt und können diese für unsere nächste Schritte verwenden:
df.to_csv("DF_Join_final_bereinigt.csv", index=False)
```
