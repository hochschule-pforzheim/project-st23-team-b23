##Forschungsfrage und Hintergrund unseres Projekts

**Research Question:** Wie erfolgreich wird ein Film bei den Konsumenten abschneiden?

*Unterfrage*: Welche Merkmale sollte ein Film erfüllen, um erfolgreich zu sein? 

Wo könnte unser Modell Anwendung finden?
Drehbuchautoren könnten durch unser Modell prüfen, wie erfolgreich deren Film in Zukunft abschneiden wird.

Unser Projekt stützt sich auf Daten des Streaming-Anbieters **Netflix**, **IMdb Ratings** und **TMBd Scores**

Hierbei werden wir den Erfolg des Films anhand des Imdb Score bzw. Tmdb Scores festlegen

Es handelt sich um eine **Regression**: Mit welchem Rating wird der Film bewertet werden?
Mit einem Regressionsmodell können wir eine Vorhersage treffen, mit welchem ImdB bzw. TMBd Rating dieser bewertet wird. Wir werden bei dem Imdb Score/tmdb Score einen gewichteten Mittelwert berechnen und die Percentile anschauen, um einen Wert festzulegen, welcher aussagt, dass ein Film als erfolgreich gilt.
