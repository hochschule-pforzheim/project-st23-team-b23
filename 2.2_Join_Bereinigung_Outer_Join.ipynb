#Bereinigung des zusammengefügten Datensatzes

Nach dem wir alle Datensätze gejoint haben und verschiedene Joins ausprobiert bereinigen wir den über den Outer Join zusammengefügten Datensatz. Dieser Join hat am besten funktioniert, die anderen hatten nur noch wenige Zeile oder viele fehlende Werte.

```python
#Daten laden

import pandas as pd

df = pd.read_csv('Join_5_movie_meta_data.csv')
```

```python
df
```

```python
#Prüfen nach fehlenden Werten

df.isna().sum()
```

Als erstes werden wir einige Spalten, die wir nicht brauchen löschen

```python
df = df.drop(['id_x', 'imdb_id_x', 'Unnamed: 0', 'imdb_id_y'], 1)
```

Es gibt viele fehlende Werte in dem Datensatz, die durch das zusammenfügen der Datensätze entstanden sind.

```python
df.isna().sum()
```

```python
#Zunächst löschen wir die fehlenden Werte in der title Spalte und der imdb und tmdb score Spalten, da diese für unsere Modell am wichitgsten sind

df = df.dropna(subset = ['title', 'imdb_score', 'tmdb_score'], axis=0)
```

```python
df.isna().sum()
```

```python
df
```

```python
#Um auf einen Datensatz ohne fehlende Werte zu kommen müssen wir noch einige Spalten löschen, in denen es nur wenige Werte gibt

df = df.drop(['budget_x', 'genres_y', 'id_y', 'original_language', 'production_companies', 'production_countries_y', 'tagline', 'awards', 'producers', 'budget_y', 'production companies', 'plot', 'plot outline', 'tageline'], 1)
```

```python
df.isna().sum()
```

```python
#Fehlende Werte in Prozent
(df.isna().sum()/len(df)).apply('{0:.4%}'.format)
```

```python
#Wir haben immer noch sehr viel fehlende Werte, wenn wir alle bereinigen hat unser Datensatz keine Zeilen mehr

df_clean = df.dropna()

df_clean
```

```python
#Daher müssen wir leider noch mehr Spalten entfernen

df = df.drop(['Academy Awards adapted screenplay', 'Academy Awards original screenplay', 'BAFTA nominations', 'Golden Globe Award for Best Screenplay', 'Writers Guild Awards Winners & Nominees 2020-2013', 'director', 'cast', 'country', 'genres'],1)
```

```python
df.isna().sum()
```

```python
df = df.dropna()
```

```python
df.isna().sum()
```

```python
df
```

```python
#Nach dem wir die Nullwerte bereinigt haben wollen wir noch nach Duplikaten suchen

df.duplicated().sum()
```

```python
#Da es nur 3 Zeilen sind können wir diese entfernen

df = df.drop_duplicates()
```

```python
#Durch die Angabe der Schauspieler sind die Titel mehrmals enthalten, wir wollen sehen wie viele Filme in dem Datensatz enthalten sind:

uniques = pd.unique(df[['title']].values.ravel())

len(uniques)
```

```python
df
```

Wir erhalten einen Datensatz mit 3950 Filmen. Wenn wir diesen Datensatz mit den ursprünglichen Daten vergleichen sehen wir, dass er sehr ähnlich ist zu den Datensatz credits und titles und der Join der vielen Datensätze keinen Mehrwer. Daher würde wir im nächsten Schritt nur diese beiden Datensätze joinen um eine solide Datenbasis für unser Modell zu erhalten.

```python
#Datei abspeichern:

df.to_csv("Df_Join_5_bereinigt.csv", index=False)
```
