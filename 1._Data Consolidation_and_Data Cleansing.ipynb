## Cleaning tmdb_5000_movies.csv

### Spalten aufbereiten

Wir entfernen alle Spalten, die nicht relevant für unser Projket sind. Bevor wir mit der Bereinigung gestartet haben, haben wir uns alle Datensätze angeschaut und miteinander verglichen. Dabei haben wir uns eine Liste erstellt, welche Spalten nicht relevant für unser Model sein werden. Außerdem haben wir notiert, welche Spalten umbenannt werden müssen bzw. welche Datentypen umgewandelt werden müssen, damit alle Datensätze identisch sind und gejoint werden können. Daher muss an dieser Stelle auch nur der "original title" behalten werden, da dieser auch in den anderen Datensätzen enthalten ist. Jedoch benennen wir die Spalte in "title" um, damit alle Spalten aller Datensätze gleich benannt sind. Die ursprüngliche Spalte "title" kann entfernt werden.

```python
#Entfernen von Spalten, die nicht gebraucht werden
df_tmdb_5000_clean = (df_tmdb_5000.
                      drop(['homepage','keywords', 'overview', 'popularity', 'release_date', 'revenue', 'spoken_languages', 'status', 'title', 'vote_count'], 1))
```

### Datentypen anpassen

```python
#Datentypen prüfen
df_tmdb_5000_clean.dtypes
```

```python
#budget in object umwandeln
df_tmdb_5000_clean2 = df_tmdb_5000_clean.astype({'budget': object})
```

```python
#id in object umwandeln
df_tmdb_5000_clean3 = df_tmdb_5000_clean2.astype({'id': object})
```

```python
#Datentypen prüfen, ob nun korrekt
df_tmdb_5000_clean3.dtypes
```

```python
#Spalten umbenennen: "original_title" und "vote_average"
df_tmdb_5000_clean3.rename(columns={'original_title': 'title', 'vote_average': 'tmdb_score'}, inplace=True)
```

```python
#Überprüfen, ob alle columns nun korrekt heißen
df_tmdb_5000_clean3
```

### Speichern des bereinigten Datensatzes

```python
#Speichern der gereinigten Daten als csv-Datei
df_tmdb_5000_clean3.to_csv("tmdb_5000_join.csv", index=False)
```
